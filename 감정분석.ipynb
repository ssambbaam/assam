```python
from transformers import pipeline
import numpy as np

# 한국어 감정분석 모델
try:
    model1 = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')
    model2 = pipeline('sentiment-analysis', model='snunlp/KR-FinBERT')
    model3 = pipeline('sentiment-analysis', model='monologg/koelectra-base-v3-discriminator')
except Exception as e:
    print(f"모델 로드 오류: {e}")
    exit()

def analyze_sentiment(text):
    # 각각 감정분석 수행
    result1 = model1(text)[0]
    result2 = model2(text)[0]
    result3 = model3(text)[0]

    # 결괏값
    scores = [result1['score'], result2['score'], result3['score']]
    labels = [result1['label'], result2['label'], result3['label']]

    print([scores])

    # 앙상블 예측 수행
    ensemble_score = np.mean(scores)
    ensemble_label = labels[np.argmax(scores)]

    # 감정 라벨
    if ensemble_score <= 0.2:
        ensemble_label = '매우 부정'
    elif ensemble_score <= 0.4:
        ensemble_label = '부정'
    elif ensemble_score <= 0.6:
        ensemble_label = '중립'
    elif ensemble_score <= 0.8:
        ensemble_label = '긍정'
    else:
        ensemble_label = '매우 긍정'

    return ensemble_score, ensemble_label

text = input("텍스트를 입력해주세요: ")
score, label = analyze_sentiment(text)
print(f"감정 점수: {score:.2f}")
print(f"감정 라벨: {label}")
```
